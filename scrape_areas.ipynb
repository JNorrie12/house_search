```python
from rightmove_scraper import *
from areas_config import london_areas, non_london_areas, area_csv_path
```

```python
today = pd.Timestamp('today').date()
```

```python
scraper = RightmoveScraper()

for area, area_url in areas.items():

    logger.info(f"Searching for properties in {london_areas}")

    properties = scraper.search_properties(area_url)

    if properties:
        logger.info(f"Found {len(properties)} properties")

        # Save results to CSV
        df = pd.DataFrame(properties)
        df.to_csv(area_csv_path(area, today), index=False)
        logger.info(f"Results saved to {area_csv_path(area, today)}")
    else:
         logger.warning("No properties found")
```

```python
scraper = RightmoveScraper()

for area, area_url in non_london_areas.items():

    logger.info(f"Searching for properties in {area}")

    properties = scraper.search_properties(area_url)

    if properties:
        logger.info(f"Found {len(properties)} properties")

        # Save results to CSV
        df = pd.DataFrame(properties)
        df.to_csv(area_csv_path(area, today), index=False)
        logger.info(f"Results saved to {area_csv_path(area, today)}")
    else:
         logger.warning("No properties found")
```
